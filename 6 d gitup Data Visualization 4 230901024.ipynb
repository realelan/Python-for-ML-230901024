```python
#Seaborn Scatter Plot and Bar Plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px

#seaborn package
Xpoints=np.array([0,20,120,200])
Ypoints=np.array([10,80,100,120,150,180,250])
print(sns.scatterplot(data=Ypoints))
print(plt.show())

# Seaborn bar plot
print(sns.barplot(data=Ypoints))
print(plt.show())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 3D Plot
X = np.array([1, 2, 6, 8, 12, 14, 20])
Y = np.array([3, 8, 1, 10, 12, 16, 18])

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(X, Y, Y)  # Adjusted the axes for 3D plotting
plt.colorbar(ax.scatter3D(X, Y, Y))  # Color bar
plt.show()
```

```python
import plotly.express as px
import numpy as np

# Plotly scatter plot
X = np.array([10, 80, 100, 120, 150, 180, 250])

fig = px.scatter(x=X)
fig.show()
```
